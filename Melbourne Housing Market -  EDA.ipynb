## Some Key Details

**Suburb**: Suburb

**Address:** Address  

**Rooms:** Number of rooms  

**Price:** Price in Australian dollars  

**Type:**  
br - bedroom(s);  
h - house,cottage,villa, semi,terrace;  
u - unit, duplex;  
t - townhouse;  
dev site - development site;

**Method:**  
S - property sold;  
SP - property sold prior;  
PI - property passed in;  
PN - sold prior not disclosed;

**Date:** Date sold

> This value makes no sense. The founding year of Melbourne was 1835 - 1888

```python
df_modif = df.copy() #creating a copy, so as not to modify the original df
print(df_modif.shape[0])
df_modif = df_modif[df_modif.YearBuilt > 1888] 
df_modif = df_modif[df_modif.YearBuilt < 2018] # only valid values 
print(df_modif.shape[0])
```

```python
# Using the categories created on the mean of the grouped dataset into the standard dataset
cat = []
for i in df_reduced['Type']:
    for c, v in dictPostCodeCat.items():
        if c == i:
            cat.append(v)
df_reduced['TypeCat'] = cat
```
